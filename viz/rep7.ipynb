```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import chart_studio.plotly as py
# Data
data = {
    'wine_names': [
        'Cabernet Sauvignon',
        'Special Selection Cabernet Sauvignon',
        'Cabernet Sauvignon',
        'Cabernet Sauvignon (Signature)',
        'CASK 23 Cabernet Sauvignon'
    ],
    'ratings_average': [4.8, 4.7, 4.6, 4.6, 4.6],
    'ratings_count': [2941, 41236, 157944, 13730, 9464]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(
    x='ratings_average',
    y='wine_names',
    data=df,
    palette='Blues_d',
    edgecolor='black'
)

# Adding a second axis for ratings count
for index, value in enumerate(df['ratings_count']):
    plt.text(df['ratings_average'][index] + 0.02, index, f'Count: {value}', va='center')

plt.title('Top 5 Cabernet Sauvignon Wines Based on Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Wine Name')
plt.xlim(4.5, 5.0)
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
data = {
    'wine_names': [
        'Cabernet Sauvignon',
        'Special Selection Cabernet Sauvignon',
        'Cabernet Sauvignon',
        'Cabernet Sauvignon (Signature)',
        'CASK 23 Cabernet Sauvignon'
    ],
    'ratings_average': [4.8, 4.7, 4.6, 4.6, 4.6],
    'ratings_count': [2941, 41236, 157944, 13730, 9464]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='ratings_average',
    y='ratings_count',
    hue='wine_names',
    data=df,
    palette='Dark2',
    s=100,  # Size of points
    edgecolor='black'
)

# Adding annotations
for i in range(df.shape[0]):
    plt.text(
        df['ratings_average'][i] + 0.01,
        df['ratings_count'][i] + 500,
        df['wine_names'][i],
        fontsize=9
    )

plt.title('Top 5 Cabernet Sauvignon Wines')
plt.xlabel('Average Rating')
plt.ylabel('Ratings Count')
plt.xlim(4.5, 5.0)
plt.ylim(0, max(df['ratings_count']) + 5000)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
data = {
    'wine_names': [
        'Cabernet Sauvignon',
        'Special Selection Cabernet Sauvignon',
        'Cabernet Sauvignon',
        'Cabernet Sauvignon (Signature)',
        'CASK 23 Cabernet Sauvignon'
    ],
    'ratings_average': [4.8, 4.7, 4.6, 4.6, 4.6],
    'ratings_count': [2941, 41236, 157944, 13730, 9464]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='ratings_average',
    y='ratings_count',
    hue='wine_names',
    data=df,
    palette='Dark2',
    s=150,  # Size of points
    edgecolor='black'
)

# Adding annotations with wine names and ratings count
for i in range(df.shape[0]):
    plt.text(
        df['ratings_average'][i] + 0.01,
        df['ratings_count'][i] + 1500,
        f"{df['wine_names'][i]} (rating_count: {df['ratings_count'][i]})",
        fontsize=10
    )

plt.title('Top 5 Cabernet Sauvignon Wines')
plt.xlabel('Average Rating')
plt.ylabel('Ratings Count')
plt.xlim(4.5, 5.0)
plt.ylim(0, max(df['ratings_count']) + 10000)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Data
data = {
    'wine_names': [
        'Cabernet Sauvignon',
        'Special Selection Cabernet Sauvignon',
        'Cabernet Sauvignon',
        'Cabernet Sauvignon (Signature)',
        'CASK 23 Cabernet Sauvignon'
    ],
    'ratings_average': [4.8, 4.7, 4.6, 4.6, 4.6],
    'ratings_count': [2941, 41236, 157944, 13730, 9464]
}


df = pd.DataFrame(data)

fig = px.scatter(
    df,
    x='ratings_average',
    y='ratings_count',
    text=df['wine_names'] ,
    
    
 
    size='ratings_count',
    color='wine_names',
    hover_name='wine_names',
    
    labels={
        
        'ratings_average': 'Average Rating',
        'ratings_count': 'Ratings Count'
    },
    title='Top 5 Cabernet Sauvignon Wines',
    color_discrete_sequence=px.colors.qualitative.Dark2, 
    size_max=50 
    
)


fig.update_traces(textposition='middle right')


fig.update_layout(
    xaxis=dict(range=[4.5, 5.0]),
    yaxis=dict(range=[0, 170000]),
                    #   max(df['ratings_count']) + 10000]),
                    
    width=700,
    height=670,
    showlegend=False,
    title_x=1,  
)
fig.update_xaxes(tickfont=dict(size=10),title_text='Average Rating')
fig.update_yaxes(tickfont=dict(size=10),title_text='Ratings Count')

fig.show()
```
